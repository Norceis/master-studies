```python
import pandas as pd

data = pd.read_csv('mieszkania.csv', delimiter=';')
data
```

```python
data = data[data.Wartosc != '0']
data = data[data.Wartosc != '-']

data.drop(columns='Atrybut', inplace=True)
data
```

```python
numeric = pd.to_numeric(data.Wartosc)
data = pd.concat([data.drop(columns='Wartosc'), numeric], axis=1)
```

```python
data.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (15,15)
# sns.displot(data=data, x='Wartosc', hue='Transakcje rynkowe', kind='kde')
sns.histplot(data=data, x='Wartosc', hue='Transakcje rynkowe', kde=True, stat='probability')
plt.show()
```

Ceny mieszkań na rynku pierwotnym są zauważalnie większe od cen mieszkań na rynku wtórnym, ale jednocześnie transakcji na rynku pierwotnym jest mniej, więc transakcje rynkowe ogółem są w dużej części transakcjami na rynku wtórnym.

1. Wyświetl funkcję masy prawdopodobieństwa dla dwóch zmiennych (`x` i `y`): `MEDV`, `RM`. W jakiej cenie i liczbie pokoi sprzedaje się najwięcej mieszkań? Co jeszcze można odczytać z wykresu?
1. Sprawdź jeszcze jeden dowolny atrybut i opisz wyniki.


```python
boston = pd.read_csv('boston.txt', skiprows=22, header=None, delimiter=r"\s+")
boston[11] = ''
boston[12] = ''
boston[13] = ''

for row in range(1,len(boston),2):
    boston.at[row-1, 11] = boston.loc[row, 0]
    boston.at[row-1, 12] = boston.loc[row, 1]
    boston.at[row-1, 13] = boston.loc[row, 2]
    boston = boston.drop(row, axis=0)

boston.reset_index(inplace=True, drop=True)
boston.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
```

```python
boston
```

```python
sns.histplot(data=boston, x='MEDV', kde=True, bins=100, stat='probability')
plt.show()
```

```python
sns.histplot(data=boston, x='MEDV', y='RM', kde=True,  stat='count', legend=True)
plt.show()
```

Najwięcej sprzedaje się mieszkań o wartości 20k dolarów i liczbie pokoi = 6. Od tego punktu stopniowo w każdym kierunku na obydwu osiach liczba wystąpień takiej kombinacji wartości i liczby pokoi stopniowo spada, ale jednoznacznie widać skupienie częstszych ofert z zakresu $15-25k i 5-7 pokoi. Widać również tendencję do układania się danych w linii prostej (przy zwiększaniu się liczby pokoi rośnie cena w sposób liniowy), ale częstość ich występowania poza podanym przeze mnie wyżej zakresem jest dużo mniejsza niż danych z podanego zakresu.

```python
sns.histplot(data=boston, x='MEDV', y='AGE', kde=True,  stat='count', legend=True)
plt.show()
```

Najwięcej sprzedaje się mieszkań starych (wybudowanych przed 1940) o stosunkowo małej wartości (ok. 15k dolarów). Widać również prawidłowość - nowsze mieszkania (stosunek ilości budynków wybudowanych po 1940 i przed) mieszczą się w innym zakresie cenowym niż te starsze (rzędu $20-35k). Największa sprzedawana ilość mieszkań to grupa mieszkań starszych, ponieważ takich jest w zbiorze danych więcej i różnią się między sobą innymi cechami niż wiek, więc ich cena jest bardzo różnorodna.
